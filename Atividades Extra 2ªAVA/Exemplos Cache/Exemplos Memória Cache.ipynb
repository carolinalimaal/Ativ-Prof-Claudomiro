```python
"""(5.2) Cache com capacidade de 32KB
Com linhas de 8bytes de largura, ou seja, os blocos da MP têm 8bytes também = 8 células por linha ou bloco
MP com capacidade de 16MB"""

# Para calcular o número de células da MP, fazemos capacidade da MP/tam de cada célula 
celulas = (2**24) / 2**0 #2^24 células -> 24 bits para endereçamento 
print(f'{celulas} = 2^24 células -> 24 bits para endereçamento.')

# Para calcular a quantidade de bits da células, temos: 1 célula = 1byte = 8bits = 2^3 -> 3bits
print('1 célula = 1 byte = 8bits = 2^3 -> 3bits para célula.')

# Para calcular o número de linhas, fazemos capacidade da Cache/nº células na linha
linhas = (2**15) / (2**3) #2^12 -> 12bits para linha
print(f'{linhas} = 2^12 linhas -> 12bits para linha.')

# Para calcular o número de blocos, fazemos total células da MP/nº células o bloco 
blocos = (2**24) / (2**3) #2^21 blocos 
print(f'{blocos} = 2^21 blocos.')

# Para calcular a Tag, fazemos nº blocos/nº linhas 
tag = 2**21 / 2**12 #2^9 blocos por linha -> 9 bits para Tag
print(f'{tag} blocos por linha = 2^9 -> 9 bits para Tag.')

# A quantidade de bits da cache = quant. bits para armaz. + quant linhas * quant bits para identificar tags

bits_cache = 32*1000*8 + 4*1000*9
print(f'RESPOSTA: {bits_cache} bits na cache.')
```

```python
"""(5.3)Capacidade MP 64MB
CACHE com 2K linhas, cada uma com largura de 16bytes"""

# Para calcular o total de células da MP, fazemos capacidade da MP/tam da célula
tot_celulas = 2**26/2**0 
print(f'{tot_celulas} = 2^26 células -> 26 bits de endereço.')

# 1 Linha com 16Bytes = 1 Bloco com 16 bytes -> cada bloco tem 16 células
print('16 células = 2^4 -> 4bits para célula.') 

# Para calcular o número de blocos, fazemos total células/nº células por bloco
blocos = 2**26 / 2**4 
print(f'{blocos} = 2^22 blocos.')

# Para saber quantos bits serão necessários para as linhas, temos 
print('2K = 2^11 linhas -> 11 bits para linhas.')

# Para calcular a tag, fazemos nº blocos/nº linhas
tag = 2**22/2**11 
print(f'{tag} = 2**11 -> 11 bits para tag.')

print('RESPOSTA: 11 bits para tag, 11bits para linha, 4bits para célula = 26bits de endereço.')
```

```python
"""5.8)Cache com 32KB, com linhas de 8bytes e conjuntos de 4 linhas
MP com 16MB"""

print('8bytes = 8 células por linha/bloco = 2^3 -> 3bits para células.')

#Para calcular nº células, fazemos capacidade MP/tam célula
celulas = 2**24/2**0
print(f'{celulas} = 2^24 células -> 24bits para endereço.')

#Para calcular o nº blocos, fazemos nº células/nº células por bloco
blocos = 2**24/2**3
print(f'{blocos} = 2^21 blocos.')

#Para calcular o nº de linhas, fazemos capacidade cache/nº células por linha
linhas = 2**15/2**3
print(f'{linhas} = 2^12 linhas.')

#Para calcular o nº de conjuntos, fazemos nº linhas/nº de linhas por conjunto
conjuntos = 2**12/2*2
print(f'{conjuntos} = 2^10 conjuntos -> 10bits para conjunto.')

#Para calcular a tag, fazemos nº blocos/nº conjuntos
tag = 2**21/2**10
print(f'{tag} = 2^11 blocos por conjunto -> 11bits para tag.')

#Quantidade de bits da cache = quant bits de armaz. + quant linhas * quant bits para identificar tags
bits_cache = 32*1000*8 + 4*1000*11
print(f'RESPOSTA: {bits_cache} bits da cache')
```
