```python
"""(5.2) Cache com capacidade de 32KB
Com linhas de 8bytes de largura, ou seja, os blocos da MP tm 8bytes tambm = 8 clulas por linha ou bloco
MP com capacidade de 16MB"""

# Para calcular o nmero de clulas da MP, fazemos capacidade da MP/tam de cada clula 
celulas = (2**24) / 2**0 #2^24 clulas -> 24 bits para endereamento 
print(f'{celulas} = 2^24 clulas -> 24 bits para endereamento.')

# Para calcular a quantidade de bits da clulas, temos: 1 clula = 1byte = 8bits = 2^3 -> 3bits
print('1 clula = 1 byte = 8bits = 2^3 -> 3bits para clula.')

# Para calcular o nmero de linhas, fazemos capacidade da Cache/n clulas na linha
linhas = (2**15) / (2**3) #2^12 -> 12bits para linha
print(f'{linhas} = 2^12 linhas -> 12bits para linha.')

# Para calcular o nmero de blocos, fazemos total clulas da MP/n clulas o bloco 
blocos = (2**24) / (2**3) #2^21 blocos 
print(f'{blocos} = 2^21 blocos.')

# Para calcular a Tag, fazemos n blocos/n linhas 
tag = 2**21 / 2**12 #2^9 blocos por linha -> 9 bits para Tag
print(f'{tag} blocos por linha = 2^9 -> 9 bits para Tag.')

# A quantidade de bits da cache = quant. bits para armaz. + quant linhas * quant bits para identificar tags

bits_cache = 32*1000*8 + 4*1000*9
print(f'RESPOSTA: {bits_cache} bits na cache.')
```

```python
"""(5.3)Capacidade MP 64MB
CACHE com 2K linhas, cada uma com largura de 16bytes"""

# Para calcular o total de clulas da MP, fazemos capacidade da MP/tam da clula
tot_celulas = 2**26/2**0 
print(f'{tot_celulas} = 2^26 clulas -> 26 bits de endereo.')

# 1 Linha com 16Bytes = 1 Bloco com 16 bytes -> cada bloco tem 16 clulas
print('16 clulas = 2^4 -> 4bits para clula.') 

# Para calcular o nmero de blocos, fazemos total clulas/n clulas por bloco
blocos = 2**26 / 2**4 
print(f'{blocos} = 2^22 blocos.')

# Para saber quantos bits sero necessrios para as linhas, temos 
print('2K = 2^11 linhas -> 11 bits para linhas.')

# Para calcular a tag, fazemos n blocos/n linhas
tag = 2**22/2**11 
print(f'{tag} = 2**11 -> 11 bits para tag.')

print('RESPOSTA: 11 bits para tag, 11bits para linha, 4bits para clula = 26bits de endereo.')
```
