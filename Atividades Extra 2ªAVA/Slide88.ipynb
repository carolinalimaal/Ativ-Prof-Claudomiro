BARRAMENTO DE CONTROLE:  ele quem vai conduzir os comandos de WRITE, READ e WAIT entre processador e memria principal.  bidirecional. 

BARRAMENTO DE ENDEREOS:  ele quem pega o endereo solicitado no processador e leva at o controlador da memria principal.  unidirecional.

BARRAMENTO DE DADOS:  ele quem vai levar os dados do processador para a memria (WRITE) e da memria para o processador (READ).  bidirecional.

```python
from IPython.display import Image
Image(filename='Operao de leitura.png')
```

Operao de leitura:
1. O Registrador de endereos de memria (REM) recebe de outro registrador do processador qual o endereo da memria que ser lido.
2. O endereo  encaminhado pela UC para o Barramento de endereos (BE).
3. A unidade de controle (UC) emite o comando de Leitura para o Barramento de controle (BC).
4. O controlador de memria decodifica o comando e localiza a clula do endereo desejado.
5. O contedo da clula  encaminhado para o Barramento de dados (BD)
6. O Registrador de dados de memria (RDM) recebe pelo BD o contedo da clula, que ser encaminhado para outro registrador do processador

```python
from IPython.display import Image
Image(filename='Operao de escrita.png')
```

Operao de escrita:

1. O Registrador de endereo de memria (REM) recebe de outro registrador do processador o endereo no qual ser escrito.
2. O endereo desejado  encaminhado pela UC para o Barramento de endereos (BE).
3. O Registrador de dados de memria (RDM) recebe de outro registrador do processador o dado o qual ser escrito na memria.
4. O dado  encaminhado pela UC para o Barramento de dados (BD).
5. A Unidade de controle (UC) emite o comado de Escrita para o Barramento de controle (BC).
6. O controlador de memria decodifica o comando, localiza o endereo da clula e tranfere os dados do BD para a clula.


```python
tam_celula = tam_caracter = 1 # 1byte = 8bits 
print(f"""TAMANHO DE CADA CLULA: {tam_celula} byte
TAMANHO DE CADA CARACTER: {tam_caracter} byte\n""")
# Se uma clula tem o mesmo tamanho do caracter, cada caracter ocupa 1 clula 

quant_caracter = 128 
# Ento cada letra vai ocupar 128 clulas
quant_letras = 9 
# Se do 'A' at 'I' so 9 letras, ento do 1 'A' at o ltimo 'I' sero:
print(f"""QUANTIDADE DE CARACTERES: {quant_caracter}
QUANTIDADE DE LETRAS: {quant_letras}
TOTAL DE CARACTERES: {quant_caracter * quant_letras}\n""")

# O endereo inicial  27FA em hexadecimal
# Transformando para binrio seria 0010 0111 1111 1010
# Transformando para decimal seria 10234
end_inicial = 10234
print(f"""END INICIAL EM HEXADECIMAL: {hex(10234)[2::]}
END INICIAL EM BINRIO: {bin(10234)[2::]}
END INICIAL EM DECIMAL: {end_inicial}\n""")

# O endereo que iniciar o 1 'J' vai ser o endereo inicial mais a quantidade de clulas que sero ocupadas do 'A' at o 'I'
end_final = end_inicial + (quant_caracter * quant_letras)
print(f"""O endereo que o 1 'J' ser escrito ser o endereo do 1 'A' + a quantidade total de caracteres at o ltimo 'I'.
END FINAL: {hex(end_final)[2::]} """)
```

```python
#Definio portas lgicas
def notgate(input1):
    if input1 == 1:
        return 0 
    else:
        return 1
    
def andgate(input1, input2, input3):
    if input1 == input2 == input3 == 1:
        return 1
    else: 
        return 0
    
def orgate(input1, input2, input3):
    if input1 == input2 == input3 == 0:
        return 0
    else: 
        return 1

# Entradas
A2, A1, A0 = int(input('Insira [0] ou [1]: ')), int(input('Insira [0] ou [1]: ')), int(input('Insira [0] ou [1]: '))

# Decodificador 3 para 8
if andgate(notgate(A2), notgate(A1), notgate(A0)):
    print('1001')

if andgate(notgate(A2), notgate(A1), A0):
    print('0110')

if andgate(notgate(A2), A1, notgate(A0)):
    print('1100')

if andgate(notgate(A2), A1, A0):
    print('0001')

if andgate(A2, notgate(A1), notgate(A0)):
    print('1010')

if andgate(A2, notgate(A1), A0):
    print('0011')

if andgate(A2, A1, notgate(A0)):
    print('0000')

if andgate(A2, A1, A0):
    print('0100')
    
```
