BARRAMENTO DE CONTROLE: é ele quem vai conduzir os comandos de WRITE, READ e WAIT entre processador e memória principal. É bidirecional. 

BARRAMENTO DE ENDEREÇOS: é ele quem pega o endereço solicitado no processador e leva até o controlador da memória principal. É unidirecional.

BARRAMENTO DE DADOS: é ele quem vai levar os dados do processador para a memória (WRITE) e da memória para o processador (READ). É bidirecional.

```python
from IPython.display import Image
Image(filename='Operação de leitura.png')
```

Operação de leitura:
1. O Registrador de endereços de memória (REM) recebe de outro registrador do processador qual o endereço da memória que será lido.
2. O endereço é encaminhado pela UC para o Barramento de endereços (BE).
3. A unidade de controle (UC) emite o comando de Leitura para o Barramento de controle (BC).
4. O controlador de memória decodifica o comando e localiza a célula do endereço desejado.
5. O conteúdo da célula é encaminhado para o Barramento de dados (BD)
6. O Registrador de dados de memória (RDM) recebe pelo BD o conteúdo da célula, que será encaminhado para outro registrador do processador

Operação de escrita:

1. O Registrador de endereço de memória (REM) recebe de outro registrador do processador o endereço no qual será escrito.
2. O endereço desejado é encaminhado pela UC para o Barramento de endereços (BE).
3. O Registrador de dados de memória (RDM) recebe de outro registrador do processador o dado o qual será escrito na memória.
4. O dado é encaminhado pela UC para o Barramento de dados (BD).
5. A Unidade de controle (UC) emite o comado de Escrita para o Barramento de controle (BC).
6. O controlador de memória decodifica o comando, localiza o endereço da célula e tranfere os dados do BD para a célula.


```python
tam_celula = tam_caracter = 1 # 1byte = 8bits 
print(f"""TAMANHO DE CADA CÉLULA: {tam_celula} byte
TAMANHO DE CADA CARACTER: {tam_caracter} byte\n""")
# Se uma célula tem o mesmo tamanho do caracter, cada caracter ocupa 1 célula 

quant_caracter = 128 
# Então cada letra vai ocupar 128 células
quant_letras = 9 
# Se do 'A' até 'I' são 9 letras, então do 1º 'A' até o último 'I' serão:
print(f"""QUANTIDADE DE CARACTERES: {quant_caracter}
QUANTIDADE DE LETRAS: {quant_letras}
TOTAL DE CARACTERES: {quant_caracter * quant_letras}\n""")

# O endereço inicial é 27FA em hexadecimal
# Transformando para binário seria 0010 0111 1111 1010
# Transformando para decimal seria 10234
end_inicial = 10234
print(f"""END INICIAL EM HEXADECIMAL: {hex(10234)[2::]}
END INICIAL EM BINÁRIO: {bin(10234)[2::]}
END INICIAL EM DECIMAL: {end_inicial}\n""")

# O endereço que iniciará o 1º 'J' vai ser o endereço inicial mais a quantidade de células que serão ocupadas do 'A' até o 'I'
end_final = end_inicial + (quant_caracter * quant_letras)
print(f"""O endereço que o 1º 'J' será escrito será o endereço do 1º 'A' + a quantidade total de caracteres até o último 'I'.
END FINAL: {hex(end_final)[2::]} """)
```

```python
#Definição portas lógicas
def notgate(input1):
    if input1 == 1:
        return 0 
    else:
        return 1
    
def andgate(input1, input2, input3):
    if input1 == input2 == input3 == 1:
        return 1
    else: 
        return 0
    
def orgate(input1, input2, input3):
    if input1 == input2 == input3 == 0:
        return 0
    else: 
        return 1

# Entradas
A2, A1, A0 = int(input('Insira [0] ou [1]: ')), int(input('Insira [0] ou [1]: ')), int(input('Insira [0] ou [1]: '))

# Decodificador 3 para 8
if andgate(notgate(A2), notgate(A1), notgate(A0)):
    print('1001')

if andgate(notgate(A2), notgate(A1), A0):
    print('0110')

if andgate(notgate(A2), A1, notgate(A0)):
    print('1100')

if andgate(notgate(A2), A1, A0):
    print('0001')

if andgate(A2, notgate(A1), notgate(A0)):
    print('1010')

if andgate(A2, notgate(A1), A0):
    print('0011')

if andgate(A2, A1, notgate(A0)):
    print('0000')

if andgate(A2, A1, A0):
    print('0100')
    
```
